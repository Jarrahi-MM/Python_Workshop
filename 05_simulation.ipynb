```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, t
import warnings
warnings.filterwarnings('ignore')
```

```python
np.random.normal(loc=8, scale=5, size=(4,2))
```

```python
DIST_MEAN, DIST_SD = 10, 3
N = 5
M = 1_000_000
d = np.random.normal(loc=DIST_MEAN, scale=DIST_SD, size=(M,N))
print(d[:3])
```

```python
s = d.mean(axis=1)
print(s.shape)
print(s[:3])
```

```python
print(f'Distribution mean is {DIST_MEAN}')
print(f'Sample Mean is {s.mean()}')
```

```python
print(f'Distribution variance is {DIST_SD**2 / N}')
print(f'Sample variance is {np.var(s)}')
```

## Central limit theorem

```python
DIST_MEAN, DIST_SD = 10, 3
N = 5
M = 30
s = np.random.normal(loc=DIST_MEAN, scale=DIST_SD, size=(M,N)).mean(axis=1)
sns.distplot(x = s);
```

```python
DIST_MEAN, DIST_SD = 10, 3
N = 5
M = 1000
s = np.random.normal(loc=DIST_MEAN, scale=DIST_SD, size=(M,N)).mean(axis=1)
sns.distplot(x = s);
```

```python
DIST_MEAN, DIST_SD = 10, 3
N = 5
M = 100_000
s = np.random.normal(loc=DIST_MEAN, scale=DIST_SD, size=(M,N)).mean(axis=1)
sns.distplot(x = s);
```

```python
DIST_MEAN, DIST_SD = 10, 3
N = 5
M = 100
s = np.random.normal(loc=DIST_MEAN, scale=DIST_SD, size=(M,N)).mean(axis=1)
sns.distplot(x = s, hist=False)

x = np.arange(4, 16, 0.1)
y = norm.pdf(x, loc=DIST_MEAN, scale=DIST_SD/N**0.5)
sns.lineplot(x=x, y=y)
plt.legend(["simulation", "normal dist"]);
```

```python
DIST_MEAN, DIST_SD = 10, 3
N = 5
M = 100_000
s = np.random.normal(loc=DIST_MEAN, scale=DIST_SD, size=(M,N)).mean(axis=1)
sns.distplot(x = s, hist=False)

x = np.arange(4, 16, 0.1)
y = norm.pdf(x, loc=DIST_MEAN, scale=DIST_SD/N**0.5)
sns.lineplot(x=x, y=y)
plt.legend(["simulation", "normal dist"]);
```

## T-Dist vs Normal-Dist

```python
plt.xlim([-5, 5])

DIST_MEAN, DIST_SD = 10, 3
N = 3
M = 100_000
d = np.random.normal(loc=DIST_MEAN, scale=DIST_SD, size=(M,N))
s = (d.mean(axis=1) - DIST_MEAN) / (d.var(axis=1, ddof=1)/N)**0.5
s[s<-10]=-10 # technical issue
s[s>10]=10 # technical issue
sns.distplot(x = s, hist=False, label='simulation')

x = np.arange(-5, 5, 0.1)
sns.lineplot(x=x, y=norm.pdf(x), label='normal')

sns.lineplot(x=x, y=t.pdf(x, df=N-1), label='t')

plt.legend();
```

```python
plt.xlim([-5, 5])

DIST_MEAN, DIST_SD = 10, 3
N = 20
M = 100_000
d = np.random.normal(loc=DIST_MEAN, scale=DIST_SD, size=(M,N))
s = (d.mean(axis=1) - DIST_MEAN) / (d.var(axis=1, ddof=1)/N)**0.5
s[s<-10]=-10 # technical issue
s[s>10]=10 # technical issue
sns.distplot(x = s, hist=False, label='simulation')

x = np.arange(-5, 5, 0.1)
sns.lineplot(x=x, y=norm.pdf(x), label='normal')

sns.lineplot(x=x, y=t.pdf(x, df=N-1), label='t')

plt.legend();
```
